```python
#create a datataframe
country = ['Kenya', 'Rwanda', 'Burundi', 'Tanzania', 'Uganda']
Population = [2040, 6080, 6098, 5678, 4567]
countryPop = pd.DataFrame({"Country": country,
                          "Population": Population})
countryPop
```

```python
#Add a series
countryPop['size'] = pd.Series([2345, 6754, 67789, 778890, 23566])
countryPop
```

```python
#Add a series from existing columns
countryPop['GDP'] = countryPop.size / countryPop.Population
countryPop
```

```python
#Adding a series from existing data Method2
countryPop['GDP2']= countryPop['size'] * countryPop['Population']
countryPop
```

forward-fill
data.fillna(method = 'ffill') to propagate the previous values forward
data.fillna(method = 'ffill', axis = 1) to propagate the previous values forward in a dataframe(0 for rows, 1 for columns)

back-fill
data.fillna(method = 'bfill') to propagate the next values backwards

```python
agriculture = pd.DataFrame({'Country': ['Kenya', 'Rwanda', 'Burundi', 'Tanzania', 'Uganda'],
                 'Livestock': [6746, 8977, 2738, 1537, 3736],
                  'CoffeeFarms': [20, 12, 8, 7, 34],
                   'teaFarms': [23, 56, 7, 98, 23],
                    'maizeProd': [34467, 345, 34577, 8996, 3456] })
agriculture
```

```python
#merge datasets
#Pandas combines rowwise by default. To combine columnwise you need to specify the axis=1
country_data = pd.concat([countryPop, agriculture], axis=1)
country_data
```

```python
#merge data columnwise by a specific column
country_data2 = pd.merge(countryPop, agriculture, on= 'Country')
country_data2
```
