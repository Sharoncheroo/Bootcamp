```python
#clean earnings
data.earnings = data['earnings'].str.replace("$","")
```

```python
def convert(value):
  unit = value[-1].upper()
  if unit == "M":
    return float(value[:-1]) * 1000000
  elif unit == "K":
    return float(value[:-1]) * 1000
  else:
    return float(value)

# Example usage:
data['earnings'] = data['earnings'].apply(convert)

data['earnings'] = data['earnings'].astype(int)


# value = "56.7M"
# converted_value = convert(value)
# print(converted_value) 
```

```python
data.head()
```

```python
summary_table = data.groupby('country')['earnings'].mean().astype((int)).reset_index()
summary_table.columns = ['country', 'average_earnings']
summary_table = summary_table.sort_values(by='average_earnings', ascending=True)
summary_table
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 8))
ax = summary_table.plot(x='country', y='average_earnings', kind='barh')
# Define a function to format the y-axis labels (convert numbers to millions)
def millions_formatter(x, pos):
  return f"{x/1e6:.0f}M"  # Format as millions, remove decimals

# Set the y-axis formatter
ax.xaxis.set_major_formatter(millions_formatter)
plt.show()
```
