```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv")
data.head()
```

```python
data.info()
```

### Bar charts

```python
data.columns
```

```python
#Count occurrences of each region
region_count = data['region'].value_counts()

#Bar graph of the regions
plt.figure(figsize=(8,6))
region_count.plot(kind='bar')
plt.title('Bar Graph of Regions', color='red')
plt.xlabel('Geographical Region')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
crime_rate_region = data.groupby('region')['crime_rate'].sum()
crime_rate_region

#Bar graph
crime_rate_region.plot(kind='bar')
plt.title('crime rate per region')
plt.xlabel('Region')
plt.ylabel('Crime Rate')
plt.xticks(rotation =0)
plt.grid(axis='y')
plt.show()
```

## Histogram

```python
#Distribution of income - histogram
plt.figure(figsize=(8,6))
plt.hist(data['income'], edgecolor='black')
plt.title('Total Income in 1976 in millions of dollars')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()


```

### Pie Charts

```python
data.columns
```

```python
crime_rate_region = data.groupby(['region']).sum()
crime_rate_region
```

```python
total_crime_rate = data['crime_rate'].sum()
total_crime_rate
```

```python
crime_rate_region = data.groupby('region')['crime_rate'].sum()
crime_rate_region
crime_rate_percentage = (crime_rate_region/total_crime_rate)*100
crime_rate_percentage
```

```python
Explode = (0,0,0.05,0)
#crime rate % per region
plt.figure(figsize=(20,6))
plt.pie(crime_rate_percentage, labels=crime_rate_percentage.index, autopct='%1.0f%%', explode=Explode)
plt.title('Crime rate percentage per region')
plt.show()
```
